## Exercise 1: Querying the `DataMuse` API

For this exercise, you will make use of the [Datamuse API](https://www.datamuse.com/api/). The Datamuse API is a freely accessible API (without API key, for up to 100,000 requests a day). The API can be used to query for specific words, based on the constraints that you define. The documentation about the API can be found here:

[https://www.datamuse.com/api/](https://www.datamuse.com/api/)

Your objective is to consult the documentation, in order to understand how it works. You will then create a Python function called `get_datamuse_nouns_for_adjective`, that makes use the API. This function takes a specific adjective (as a string) as input. The function then queries the Datamuse API for the nouns that are most often described by the specific adjective. The function then returns the top 5 words of the result. As an example: for the adjective `yellow`, your function should return the list: `['fever', 'color', flowers', 'light', 'colour']`.

```python
#Step #1: Import required modules
import requests
import json

#Step #2 Define Function
def get_datamuse_nouns_for_adjective(adjective):
    parameters = {'rel_jja': adjective}
    response = requests.get("https://api.datamuse.com/words", params=parameters)
    results = response.json()
    nouns = [word['word'] for word in results[0:5]] #List comprehension that collects the first five nouns for chosen adjective
    if results:
        print(f"The following five nouns are frequently modified by the adjective '{adjective}': {nouns}.")
    else:
        print(f"Sorry, the Datamuse API does not contain any nouns for the adjective '{adjective}'.\n"
              f"Please ensure there are no typos or visit https://www.datamuse.com/api/ for more information.")
              #Error message letting user know they should check their word or visit the Datamuse API.

#Step #3: Execute Function
get_datamuse_nouns_for_adjective("yellow")
```
